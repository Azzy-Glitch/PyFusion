```python
# !pip install reportlab

from reportlab.lib.pagesizes import A4
from reportlab.lib import colors
from reportlab.lib.units import mm, inch
from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle
from reportlab.lib.enums import TA_CENTER, TA_LEFT, TA_RIGHT, TA_JUSTIFY
from reportlab.platypus import (
    SimpleDocTemplate, Paragraph, Spacer, Table, TableStyle,
    HRFlowable, PageBreak, KeepTogether
)
from reportlab.lib.colors import HexColor, white, black
import reportlab.lib.colors as rl_colors

OUTPUT = "QuantumLogics_Python_Curriculum.pdf"

# ── Brand Colors ──────────────────────────────────────────────────────────────
VIOLET       = HexColor("#6C3FC5")   # primary
DEEP_VIOLET  = HexColor("#4A2898")   # dark
LIGHT_VIOLET = HexColor("#EDE7FF")   # subtle bg
NEON_ACCENT  = HexColor("#A855F7")   # highlights
DARK_BG      = HexColor("#1A0A2E")   # cover bg
MINT         = HexColor("#10B981")   # success / data science
ROSE         = HexColor("#F43F5E")   # module tags
GOLD         = HexColor("#F59E0B")   # stars/awards
TEXT_DARK    = HexColor("#1E1B2E")
TEXT_MID     = HexColor("#4B5563")
TEXT_LIGHT   = HexColor("#6B7280")
ROW_ALT      = HexColor("#F5F0FF")
BORDER_LIGHT = HexColor("#D8C8F5")

W, H = A4


def make_doc():
    doc = SimpleDocTemplate(
        OUTPUT,
        pagesize=A4,
        leftMargin=18*mm, rightMargin=18*mm,
        topMargin=16*mm, bottomMargin=16*mm,
        title="Quantum Logics – Python Mastery Curriculum",
        author="Abdul Aziz | Quantum Logics Virtual Tech Academy"
    )

    styles = getSampleStyleSheet()

    # ── Custom paragraph styles ────────────────────────────────────────────────
    def S(name, **kw):
        return ParagraphStyle(name, **kw)

    cover_title   = S("CoverTitle",   fontName="Helvetica-Bold",   fontSize=36, textColor=white,         spaceAfter=4,  alignment=TA_CENTER, leading=42)
    cover_sub     = S("CoverSub",     fontName="Helvetica",         fontSize=14, textColor=HexColor("#C9B8FF"), spaceAfter=2, alignment=TA_CENTER)
    cover_meta    = S("CoverMeta",    fontName="Helvetica",         fontSize=10, textColor=HexColor("#A89CC8"), alignment=TA_CENTER, spaceAfter=3)
    cover_tagline = S("CoverTag",     fontName="Helvetica-Oblique", fontSize=12, textColor=HexColor("#E0D5FF"), alignment=TA_CENTER, spaceAfter=2)

    h_module  = S("HMod",  fontName="Helvetica-Bold", fontSize=14, textColor=white,      spaceBefore=8, spaceAfter=4, backColor=DEEP_VIOLET, leftIndent=-2, borderPad=6, alignment=TA_LEFT, leading=18)
    h_week    = S("HWeek", fontName="Helvetica-Bold", fontSize=11, textColor=DEEP_VIOLET, spaceBefore=6, spaceAfter=3, leading=14)
    h_class   = S("HClass",fontName="Helvetica-Bold", fontSize=10, textColor=VIOLET,      spaceBefore=4, spaceAfter=2, leading=13)
    h_section = S("HSection", fontName="Helvetica-Bold", fontSize=9, textColor=TEXT_DARK, spaceBefore=3, spaceAfter=1)

    body      = S("Body",  fontName="Helvetica",      fontSize=9,  textColor=TEXT_DARK,  spaceAfter=2, leading=13)
    body_sm   = S("BodySm",fontName="Helvetica",      fontSize=8,  textColor=TEXT_MID,   spaceAfter=2, leading=11)
    bullet    = S("Blt",   fontName="Helvetica",      fontSize=9,  textColor=TEXT_DARK,  leftIndent=12, spaceAfter=1, bulletIndent=4, leading=12)
    bullet_sm = S("BltSm", fontName="Helvetica",      fontSize=8,  textColor=TEXT_MID,   leftIndent=12, spaceAfter=1, bulletIndent=4, leading=11)
    label_tag = S("Tag",   fontName="Helvetica-Bold", fontSize=8,  textColor=VIOLET,     spaceAfter=1)
    center    = S("Ctr",   fontName="Helvetica",      fontSize=9,  textColor=TEXT_MID,   alignment=TA_CENTER, spaceAfter=2)
    footer_s  = S("Ftr",  fontName="Helvetica",      fontSize=7,  textColor=TEXT_LIGHT,  alignment=TA_CENTER)
    toc_entry = S("TOC",   fontName="Helvetica",      fontSize=10, textColor=TEXT_DARK,  spaceAfter=4, leading=14)
    toc_mod   = S("TOCMod",fontName="Helvetica-Bold", fontSize=11, textColor=VIOLET,     spaceAfter=2, spaceBefore=8)
    outcomes  = S("Out",   fontName="Helvetica",      fontSize=10, textColor=TEXT_DARK,  spaceAfter=3, leading=14, leftIndent=8)

    story = []

    # ══════════════════════════════════════════════════════════════════════════
    # COVER PAGE
    # ══════════════════════════════════════════════════════════════════════════
    def cover_block():
        # dark gradient-like header table
        cover_data = [[""]]
        cover_table = Table(cover_data, colWidths=[W - 36*mm], rowHeights=[260])
        cover_table.setStyle(TableStyle([
            ("BACKGROUND", (0,0), (-1,-1), DARK_BG),
            ("ROUNDEDCORNERS", [8]),
        ]))
        story.append(cover_table)

        # overlay text manually via inner table
        inner = []
        inner.append(Spacer(1, 20))
        inner.append(Paragraph("⚡ QUANTUM LOGICS", cover_sub))
        inner.append(Paragraph("VIRTUAL TECH ACADEMY", S("x", fontName="Helvetica-Bold", fontSize=11, textColor=HexColor("#8B5CF6"), alignment=TA_CENTER, spaceAfter=10)))
        inner.append(Spacer(1, 8))
        inner.append(Paragraph("Python Mastery Program", cover_title))
        inner.append(Paragraph("Zero to Advanced + Data Science Foundations", cover_sub))
        inner.append(Spacer(1, 12))
        inner.append(HRFlowable(width="60%", thickness=1, color=NEON_ACCENT, hAlign="CENTER"))
        inner.append(Spacer(1, 12))
        inner.append(Paragraph("3 Months  ·  36 Classes  ·  3 Weekend Sessions/Week  ·  90–120 Min Each", cover_tagline))
        inner.append(Spacer(1, 8))
        inner.append(Paragraph("Instructor: Abdul Aziz  |  Rawalpindi, Pakistan  |  2025", cover_meta))

        # render on top of table: rebuild as single-cell table with flowables
        cover_table2 = Table([[inner]], colWidths=[W - 36*mm])
        cover_table2.setStyle(TableStyle([
            ("BACKGROUND", (0,0), (-1,-1), DARK_BG),
            ("TOPPADDING", (0,0), (-1,-1), 28),
            ("BOTTOMPADDING", (0,0), (-1,-1), 28),
            ("LEFTPADDING", (0,0), (-1,-1), 20),
            ("RIGHTPADDING", (0,0), (-1,-1), 20),
            ("ROUNDEDCORNERS", [8]),
        ]))

        # rebuild story: remove old table
        story.clear()
        story.append(cover_table2)
        story.append(Spacer(1, 18))

        # Stats bar
        stats = [
            ["📅  3 Months", "📚  3 Modules", "🎓  36 Classes", "⏱  90–120 min"],
        ]
        stats_table = Table(stats, colWidths=[(W-36*mm)/4]*4)
        stats_table.setStyle(TableStyle([
            ("BACKGROUND", (0,0), (-1,-1), LIGHT_VIOLET),
            ("FONTNAME",   (0,0), (-1,-1), "Helvetica-Bold"),
            ("FONTSIZE",   (0,0), (-1,-1), 9),
            ("TEXTCOLOR",  (0,0), (-1,-1), DEEP_VIOLET),
            ("ALIGN",      (0,0), (-1,-1), "CENTER"),
            ("VALIGN",     (0,0), (-1,-1), "MIDDLE"),
            ("TOPPADDING", (0,0), (-1,-1), 8),
            ("BOTTOMPADDING", (0,0), (-1,-1), 8),
            ("GRID",       (0,0), (-1,-1), 0.5, BORDER_LIGHT),
            ("ROUNDEDCORNERS", [6]),
        ]))
        story.append(stats_table)
        story.append(Spacer(1, 18))

        # Program overview blurb
        blurb = (
            "This curriculum is designed by <b>Quantum Logics Virtual Tech Academy</b> to take students from "
            "absolute zero to a confident, job-ready Python developer — covering core fundamentals, "
            "object-oriented design, advanced Pythonic patterns, and a practical foundation in "
            "Data Science using NumPy, Pandas, and Matplotlib. Every session ends with a real project."
        )
        story.append(Paragraph(blurb, S("blurb", fontName="Helvetica", fontSize=10, textColor=TEXT_DARK, alignment=TA_JUSTIFY, leading=15, spaceAfter=6)))
        story.append(Spacer(1, 8))
        story.append(HRFlowable(width="100%", thickness=0.5, color=BORDER_LIGHT))
        story.append(Spacer(1, 6))

    cover_block()

    # ══════════════════════════════════════════════════════════════════════════
    # TABLE OF CONTENTS
    # ══════════════════════════════════════════════════════════════════════════
    story.append(Paragraph("Table of Contents", S("TOCHead", fontName="Helvetica-Bold", fontSize=16, textColor=VIOLET, spaceAfter=10, spaceBefore=4)))

    toc_items = [
        ("Module 1", "Python Foundations",                "Week 1–4  |  Classes 1–12",   VIOLET),
        ("  Week 1", "Introduction & Core Basics",        "Classes 1–3",                 TEXT_MID),
        ("  Week 2", "Control Flow & Collections",        "Classes 4–6",                 TEXT_MID),
        ("  Week 3", "Advanced Collections & Functions",  "Classes 7–9",                 TEXT_MID),
        ("  Week 4", "Files & Error Handling",            "Classes 10–12",               TEXT_MID),
        ("Module 2", "Intermediate Python & OOP",         "Week 5–8  |  Classes 13–24",  VIOLET),
        ("  Week 5", "OOP Basics",                        "Classes 13–15",               TEXT_MID),
        ("  Week 6", "Inheritance & Polymorphism",        "Classes 16–18",               TEXT_MID),
        ("  Week 7", "Decorators & Generators",           "Classes 19–21",               TEXT_MID),
        ("  Week 8", "Modules & Project Structure",       "Classes 22–24",               TEXT_MID),
        ("Module 3", "Data Science Foundations",          "Week 9–12  |  Classes 25–36", MINT),
        ("  Week 9",  "NumPy",                            "Classes 25–27",               TEXT_MID),
        ("  Week 10", "Pandas",                           "Classes 28–30",               TEXT_MID),
        ("  Week 11", "Data Cleaning & Visualization",    "Classes 31–33",               TEXT_MID),
        ("  Week 12", "End-to-End Data Science Pipeline", "Classes 34–36",               TEXT_MID),
        ("Appendix", "Learning Outcomes & Tools",         "",                            MINT),
    ]

    for label, title, pages, color in toc_items:
        is_mod = label.startswith("Module") or label == "Appendix"
        row_data = [[
            Paragraph(f"<b>{label}</b>" if is_mod else label,
                      S("tl", fontName="Helvetica-Bold" if is_mod else "Helvetica",
                        fontSize=10 if is_mod else 9,
                        textColor=color, leading=13)),
            Paragraph(f"<b>{title}</b>" if is_mod else title,
                      S("tt", fontName="Helvetica-Bold" if is_mod else "Helvetica",
                        fontSize=10 if is_mod else 9,
                        textColor=TEXT_DARK, leading=13)),
            Paragraph(pages, S("tp", fontName="Helvetica", fontSize=8,
                               textColor=TEXT_LIGHT, alignment=TA_RIGHT, leading=13)),
        ]]
        toc_t = Table(row_data, colWidths=[45*mm, 100*mm, 30*mm])
        toc_t.setStyle(TableStyle([
            ("BACKGROUND", (0,0), (-1,-1), LIGHT_VIOLET if is_mod else white),
            ("TOPPADDING", (0,0), (-1,-1), 5 if is_mod else 3),
            ("BOTTOMPADDING", (0,0), (-1,-1), 5 if is_mod else 3),
            ("LEFTPADDING", (0,0), (-1,-1), 8),
            ("RIGHTPADDING", (0,0), (-1,-1), 4),
            ("LINEBELOW", (0,0), (-1,-1), 0.3, BORDER_LIGHT),
        ]))
        story.append(toc_t)

    story.append(PageBreak())

    # ══════════════════════════════════════════════════════════════════════════
    # HELPER FUNCTIONS
    # ══════════════════════════════════════════════════════════════════════════

    def module_header(num, title, weeks, classes, color=DEEP_VIOLET):
        data = [[
            Paragraph(f"MODULE {num}", S("mn", fontName="Helvetica-Bold", fontSize=9,
                                         textColor=HexColor("#C9B8FF"), leading=11)),
            Paragraph(f"Weeks {weeks}  ·  {classes} Classes",
                      S("mw", fontName="Helvetica", fontSize=9, textColor=HexColor("#C9B8FF"),
                        alignment=TA_RIGHT, leading=11)),
        ], [
            Paragraph(title, S("mt", fontName="Helvetica-Bold", fontSize=16,
                               textColor=white, leading=20)),
            Paragraph("", body),
        ]]
        t = Table(data, colWidths=[(W-36*mm)*0.65, (W-36*mm)*0.35])
        t.setStyle(TableStyle([
            ("BACKGROUND",     (0,0), (-1,-1), color),
            ("TOPPADDING",     (0,0), (-1,-1), 10),
            ("BOTTOMPADDING",  (0,0), (-1,-1), 10),
            ("LEFTPADDING",    (0,0), (-1,-1), 14),
            ("RIGHTPADDING",   (0,0), (-1,-1), 10),
            ("SPAN",           (0,1), (1,1)),
            ("ROUNDEDCORNERS", [6]),
        ]))
        story.append(t)
        story.append(Spacer(1, 10))

    def week_header(week_num, week_title, color=VIOLET):
        data = [[Paragraph(f"WEEK {week_num}  —  {week_title}",
                           S("wh", fontName="Helvetica-Bold", fontSize=11,
                             textColor=white, leading=14))]]
        t = Table(data, colWidths=[W-36*mm])
        t.setStyle(TableStyle([
            ("BACKGROUND",    (0,0), (-1,-1), color),
            ("TOPPADDING",    (0,0), (-1,-1), 7),
            ("BOTTOMPADDING", (0,0), (-1,-1), 7),
            ("LEFTPADDING",   (0,0), (-1,-1), 12),
            ("RIGHTPADDING",  (0,0), (-1,-1), 8),
            ("ROUNDEDCORNERS", [4]),
        ]))
        story.append(t)
        story.append(Spacer(1, 6))

    def class_card(cls_num, cls_title, duration, concepts, micro_topics,
                   practice, table_data=None, table_headers=None, notes=None):
        """Render a single class card."""
        elements = []

        # Class title bar
        title_data = [[
            Paragraph(f"Class {cls_num}", S("cn", fontName="Helvetica-Bold", fontSize=8,
                                             textColor=NEON_ACCENT, leading=10)),
            Paragraph(f"⏱ {duration}", S("cd", fontName="Helvetica", fontSize=8,
                                          textColor=HexColor("#C9B8FF"), alignment=TA_RIGHT, leading=10)),
        ], [
            Paragraph(cls_title, S("ct", fontName="Helvetica-Bold", fontSize=11,
                                   textColor=white, leading=14)),
            Paragraph("", body),
        ]]
        class_title_t = Table(title_data, colWidths=[(W-36*mm)*0.7, (W-36*mm)*0.3])
        class_title_t.setStyle(TableStyle([
            ("BACKGROUND",    (0,0), (-1,-1), HexColor("#2D1B69")),
            ("TOPPADDING",    (0,0), (-1,-1), 6),
            ("BOTTOMPADDING", (0,0), (-1,-1), 6),
            ("LEFTPADDING",   (0,0), (-1,-1), 12),
            ("RIGHTPADDING",  (0,0), (-1,-1), 10),
            ("SPAN",          (0,1), (1,1)),
            ("ROUNDEDCORNERS", [4]),
        ]))
        elements.append(class_title_t)
        elements.append(Spacer(1, 4))

        # Body area: two columns
        left_col = []
        right_col = []

        # Concepts
        left_col.append(Paragraph("▸ CORE CONCEPTS", label_tag))
        for c in concepts:
            left_col.append(Paragraph(f"• {c}", bullet_sm))

        if micro_topics:
            left_col.append(Spacer(1, 4))
            left_col.append(Paragraph("▸ MICRO TOPICS", label_tag))
            for m in micro_topics:
                left_col.append(Paragraph(f"• {m}", bullet_sm))

        # Practice / project
        right_col.append(Paragraph("▸ HANDS-ON PRACTICE", label_tag))
        for p in practice:
            right_col.append(Paragraph(f"🔨  {p}", S("pr", fontName="Helvetica-Bold", fontSize=8,
                                                       textColor=MINT, spaceAfter=2, leading=11)))

        if notes:
            right_col.append(Spacer(1, 4))
            right_col.append(Paragraph("▸ NOTE", label_tag))
            right_col.append(Paragraph(notes, body_sm))

        body_data = [[left_col, right_col]]
        body_t = Table(body_data, colWidths=[(W-36*mm)*0.56, (W-36*mm)*0.44])
        body_t.setStyle(TableStyle([
            ("BACKGROUND",    (0,0), (0,0), ROW_ALT),
            ("BACKGROUND",    (1,0), (1,0), white),
            ("TOPPADDING",    (0,0), (-1,-1), 8),
            ("BOTTOMPADDING", (0,0), (-1,-1), 8),
            ("LEFTPADDING",   (0,0), (-1,-1), 10),
            ("RIGHTPADDING",  (0,0), (-1,-1), 10),
            ("VALIGN",        (0,0), (-1,-1), "TOP"),
            ("LINEBELOW",     (0,0), (-1,-1), 0.5, BORDER_LIGHT),
        ]))
        elements.append(body_t)

        # Optional table
        if table_data and table_headers:
            tbl = [table_headers] + table_data
            col_w = (W-36*mm) / len(table_headers)
            tbl_obj = Table(tbl, colWidths=[col_w]*len(table_headers))
            tbl_obj.setStyle(TableStyle([
                ("BACKGROUND",    (0,0), (-1,0), VIOLET),
                ("TEXTCOLOR",     (0,0), (-1,0), white),
                ("FONTNAME",      (0,0), (-1,0), "Helvetica-Bold"),
                ("FONTSIZE",      (0,0), (-1,-1), 8),
                ("FONTNAME",      (0,1), (-1,-1), "Helvetica"),
                ("TEXTCOLOR",     (0,1), (-1,-1), TEXT_DARK),
                ("ROWBACKGROUNDS",(0,1), (-1,-1), [white, ROW_ALT]),
                ("ALIGN",         (0,0), (-1,-1), "LEFT"),
                ("VALIGN",        (0,0), (-1,-1), "MIDDLE"),
                ("GRID",          (0,0), (-1,-1), 0.3, BORDER_LIGHT),
                ("TOPPADDING",    (0,0), (-1,-1), 4),
                ("BOTTOMPADDING", (0,0), (-1,-1), 4),
                ("LEFTPADDING",   (0,0), (-1,-1), 8),
                ("ROUNDEDCORNERS", [4]),
            ]))
            elements.append(Spacer(1, 4))
            elements.append(tbl_obj)

        elements.append(Spacer(1, 10))
        return elements

    def add_class(*args, **kwargs):
        story.extend(class_card(*args, **kwargs))

    # ══════════════════════════════════════════════════════════════════════════
    # MODULE 1
    # ══════════════════════════════════════════════════════════════════════════
    module_header(1, "Python Foundations", "1–4", 12, DEEP_VIOLET)

    # ── Week 1 ────────────────────────────────────────────────────────────────
    week_header(1, "Introduction & Core Basics", VIOLET)

    add_class(1, "Python Setup & Fundamentals", "90 min",
        concepts=["Interpreted vs. compiled languages — why Python runs differently",
                  "Python use cases: AI, Web Dev, Automation, Scripting",
                  "Installing Python 3.x and setting up VS Code",
                  "Running .py scripts from terminal and IDE",
                  "Understanding the Python REPL"],
        micro_topics=["print() — basic output", "Comments: # single-line, ''' multi-line '''",
                      "Indentation rules — Python's block syntax", "Python keywords overview"],
        practice=["'Hello World' → personalized output with user's name",
                  "Explore the REPL: arithmetic, strings, print()"])

    add_class(2, "Variables & Data Types", "90 min",
        concepts=["Variables as labels to memory objects",
                  "Dynamic typing — types inferred at runtime",
                  "type(), id() — inspecting type and identity",
                  "Type casting: int(), str(), float(), bool()"],
        micro_topics=[],
        practice=["Build a Simple Calculator: accepts 2 numbers, performs +/−/×/÷",
                  "Experiment with type casting and edge cases (e.g., int('abc'))"],
        table_headers=["Type", "Description", "Example"],
        table_data=[["int", "Whole integer", "age = 21"],
                    ["float", "Decimal number", "pi = 3.14"],
                    ["str", "Text string", "name = 'Aziz'"],
                    ["bool", "True / False", "active = True"],
                    ["NoneType", "Null / no value", "result = None"]])

    add_class(3, "Operators & Strings In-Depth", "90–120 min",
        concepts=["Arithmetic operators: +, −, *, /, %, //, **",
                  "Comparison operators: ==, !=, >, <, >=, <=",
                  "Logical operators: and, or, not",
                  "Strings: indexing (positive & negative), slicing, immutability",
                  "f-strings for clean formatted output"],
        micro_topics=["String concatenation vs. f-strings",
                      "Escape characters: \\n, \\t, \\\\",
                      "len(), in, not in"],
        practice=["Password Strength Checker — uses length check, digit check, upper/lower"],
        table_headers=["Method", "Purpose", "Example"],
        table_data=[["upper()", "Convert to uppercase", "'hello'.upper()"],
                    ["lower()", "Convert to lowercase", "'HELLO'.lower()"],
                    ["strip()", "Remove whitespace", "' hi '.strip()"],
                    ["replace()", "Substitute substring", "'a,b'.replace(',','/')"],
                    ["split()", "Split into list", "'a b'.split()"],
                    ["find()", "Index of substring", "'hello'.find('e')"]])

    # ── Week 2 ────────────────────────────────────────────────────────────────
    week_header(2, "Control Flow & Collections", VIOLET)

    add_class(4, "Conditionals & Decision Logic", "90 min",
        concepts=["if / elif / else syntax and nesting",
                  "Truthy and falsy values — understanding Python's boolean coercion",
                  "Ternary (one-line) conditionals",
                  "Short-circuit evaluation with and / or"],
        micro_topics=["Avoid deep nesting — guard clauses pattern",
                      "is vs. == — identity vs. equality"],
        practice=["BMI Calculator — categorize underweight / normal / overweight / obese",
                  "Grade classifier — input score, output letter grade"])

    add_class(5, "Loops & Iteration", "90 min",
        concepts=["for loop — iterating over ranges, lists, strings",
                  "while loop — condition-based repetition",
                  "Loop control: break, continue, pass",
                  "loop-else clause — runs when no break occurred",
                  "range() — start, stop, step"],
        micro_topics=["enumerate() — get index and value together",
                      "zip() — iterate multiple sequences in parallel",
                      "Nested loops — 2D iteration"],
        practice=["Number Guessing Game — random number, limited attempts, hint system"])

    add_class(6, "Lists & Tuples", "90–120 min",
        concepts=["List creation, indexing, negative indexing, slicing",
                  "Nested lists — 2D structures",
                  "List mutability vs. tuple immutability",
                  "Tuple packing and unpacking",
                  "List comprehensions — concise filtering & transformation"],
        micro_topics=["copy() vs. deepcopy() — shallow vs. deep copy",
                      "Sorting: sort() in-place vs. sorted() returns new list"],
        practice=["To-Do List System — add / remove / display / mark complete"],
        table_headers=["Method", "Function", "Example"],
        table_data=[["append(x)", "Add item to end", "lst.append(5)"],
                    ["extend([x])", "Add multiple items", "lst.extend([1,2])"],
                    ["insert(i, x)", "Insert at index", "lst.insert(0, 'a')"],
                    ["remove(x)", "Remove by value", "lst.remove(3)"],
                    ["pop(i)", "Remove by index", "lst.pop(-1)"],
                    ["sort()", "Sort in-place", "lst.sort(reverse=True)"],
                    ["reverse()", "Reverse in-place", "lst.reverse()"]])

    # ── Week 3 ────────────────────────────────────────────────────────────────
    week_header(3, "Advanced Collections & Functions", VIOLET)

    add_class(7, "Dictionaries — Mapping Data", "90 min",
        concepts=["Key-value pairs — how Python hashes keys",
                  "Dictionary creation, access, and update",
                  "Nested dictionaries for structured records",
                  "Dictionary comprehensions"],
        micro_topics=["Hashing overview — why keys must be immutable",
                      "Iterating: keys(), values(), items()",
                      "Merging dicts with ** unpacking (Python 3.9+)"],
        practice=["Contact Book — store/update/delete/search contacts by name"],
        table_headers=["Method", "Use", "Returns"],
        table_data=[["keys()", "All keys", "dict_keys"],
                    ["values()", "All values", "dict_values"],
                    ["items()", "Key-value pairs", "dict_items"],
                    ["get(k, def)", "Safe access", "Value or default"],
                    ["update(d2)", "Merge another dict", "None (in-place)"],
                    ["pop(k)", "Remove key", "Removed value"]])

    add_class(8, "Sets — Unique Collections", "90 min",
        concepts=["Sets: unordered, unique, mutable",
                  "Set creation from lists — automatic deduplication",
                  "Frozensets — immutable variant",
                  "Set algebra: union, intersection, difference, symmetric difference"],
        micro_topics=["in operator on sets — O(1) lookup",
                      "When to use sets vs. lists vs. dicts"],
        practice=["Common Interests Finder — two users enter hobbies, find shared ones",
                  "Duplicate Remover — clean a list using sets"])

    add_class(9, "Functions — The Core Abstraction", "90–120 min",
        concepts=["def keyword, parameters, return values",
                  "Positional, keyword, and default arguments",
                  "*args and **kwargs — variadic functions",
                  "Lambda functions — anonymous one-liners",
                  "Scope rules: LEGB (Local → Enclosing → Global → Built-in)",
                  "Pure functions and side effects"],
        micro_topics=["Docstrings — documenting your functions",
                      "global and nonlocal keywords",
                      "Higher-order functions: map(), filter(), sorted(key=)"],
        practice=["Utility Function Library — reusable helpers: clamp, percentage, safe_divide"])

    # ── Week 4 ────────────────────────────────────────────────────────────────
    week_header(4, "Files, Error Handling & Real Data", VIOLET)

    add_class(10, "Exception Handling", "90 min",
        concepts=["try / except / else / finally — full exception flow",
                  "Built-in exception types and when they occur",
                  "Raising custom exceptions with raise",
                  "Exception hierarchy in Python"],
        micro_topics=["Multiple except blocks for different error types",
                      "as keyword — capturing the exception object",
                      "Best practice: catch specific exceptions, not bare except"],
        practice=["Input Validation System — validate age, email, number ranges with helpful errors"],
        table_headers=["Exception", "Triggered When"],
        table_data=[["TypeError", "Wrong data type operation"],
                    ["ValueError", "Right type, wrong value"],
                    ["IndexError", "List index out of range"],
                    ["KeyError", "Dict key not found"],
                    ["ZeroDivisionError", "Dividing by zero"],
                    ["FileNotFoundError", "File path doesn't exist"]])

    add_class(11, "File Handling", "90 min",
        concepts=["open() with context manager (with statement)",
                  "File modes: r, w, a, r+, rb, wb",
                  "Reading: read(), readline(), readlines()",
                  "Writing: write(), writelines()",
                  "Working with file paths using os and pathlib"],
        micro_topics=["Why context managers (with) are safer than manual open/close",
                      "Encoding parameter — always use UTF-8"],
        practice=["Notes App — create / read / append / delete notes stored as .txt"])

    add_class(12, "CSV & JSON Data Handling", "90–120 min",
        concepts=["CSV: reading and writing tabular data with csv module",
                  "JSON: parsing and serializing structured data",
                  "json.load() / json.dump() — file operations",
                  "json.loads() / json.dumps() — string operations",
                  "Real-world use: APIs return JSON, datasets come as CSV"],
        micro_topics=["DictReader / DictWriter — working with CSV as dicts",
                      "Handling encoding and delimiter variants in CSV"],
        practice=["Student Record System — store, read, search, export student data as JSON + CSV"],
        notes="This is the capstone of Module 1. Students should build a complete mini-app.")

    story.append(PageBreak())

    # ══════════════════════════════════════════════════════════════════════════
    # MODULE 2
    # ══════════════════════════════════════════════════════════════════════════
    module_header(2, "Intermediate Python & OOP", "5–8", 12, HexColor("#7C3AED"))

    week_header(5, "Object-Oriented Programming — Fundamentals", HexColor("#7C3AED"))

    add_class("13–14", "Classes, Objects & Instance Design", "90–120 min",
        concepts=["What is OOP and why it matters in real codebases",
                  "Defining classes — class keyword, __init__, self",
                  "Instance attributes vs. class attributes",
                  "Instance methods — behavior tied to objects",
                  "__str__ and __repr__ — readable object output",
                  "Encapsulation — bundling data with behavior"],
        micro_topics=["Object identity — id(), is operator",
                      "Default mutable arguments trap in __init__",
                      "Comparing objects: __eq__, __lt__ (brief intro)"],
        practice=["Bank Account System — classes for Account, with deposit, withdraw, balance",
                  "Add overdraft protection using conditional logic"])

    add_class(15, "Class Methods, Static Methods & Properties", "90 min",
        concepts=["@classmethod — bound to class, not instance",
                  "@staticmethod — utility function inside class namespace",
                  "@property — controlled attribute access (getter pattern)",
                  "Difference: instance method vs. class method vs. static method"],
        micro_topics=["When to use each — decision framework",
                      "__slots__ — memory optimization (brief mention)"],
        practice=["Extend Bank System: add @classmethod for alternate constructors",
                  "Add @property for formatted balance display"])

    week_header(6, "Inheritance & Polymorphism", HexColor("#7C3AED"))

    add_class("16–17", "Inheritance, super() & Method Overriding", "90–120 min",
        concepts=["Single inheritance — parent → child class relationship",
                  "super() — call parent class methods cleanly",
                  "Method overriding — child specializes parent behavior",
                  "isinstance() and issubclass() — type checking at runtime",
                  "Abstract classes with abc.ABC and @abstractmethod"],
        micro_topics=["Method Resolution Order (MRO) — how Python resolves method calls",
                      "Multiple inheritance — diamond problem and how MRO solves it"],
        practice=["Animal Hierarchy — Animal → Dog / Cat / Bird with polymorphic speak() method",
                  "Employee System — Employee → Manager / Developer with specialized attributes"])

    add_class(18, "Dunder Methods & Operator Overloading", "90 min",
        concepts=["Magic methods (__add__, __len__, __contains__, __iter__)",
                  "Making custom objects work with built-in Python operators",
                  "__enter__ / __exit__ — context manager protocol",
                  "Iteration protocol: __iter__, __next__"],
        micro_topics=["When NOT to overload operators — principle of least surprise",
                      "Comparison operators: __eq__, __lt__, __le__"],
        practice=["Custom Vector class — supports +, −, dot product, len()",
                  "Build a simple context manager using __enter__/__exit__"])

    week_header(7, "Decorators & Generators", HexColor("#7C3AED"))

    add_class("19–20", "Decorators — Functions as First-Class Citizens", "90–120 min",
        concepts=["Functions are objects — passing, returning, storing functions",
                  "Closures — functions that remember their enclosing scope",
                  "@decorator syntax — syntactic sugar for wrapping",
                  "Writing reusable decorators with functools.wraps",
                  "@property, @staticmethod, @classmethod as built-in decorators",
                  "Stacking multiple decorators"],
        micro_topics=["Decorator factories — decorators that take arguments",
                      "Class-based decorators using __call__"],
        practice=["Logging Decorator — prints function name, args, return value",
                  "Timer Decorator — measures execution time",
                  "Access Control Decorator — mock role-based permission check"],
        notes="This is one of the most practically useful topics for real Python codebases.")

    add_class(21, "Generators & Iterators", "90 min",
        concepts=["Iterator protocol — __iter__, __next__",
                  "yield keyword — turning functions into generators",
                  "Generator expressions — lazy evaluation syntax",
                  "Memory efficiency: generators vs. lists at scale",
                  "Infinite sequences using generators"],
        micro_topics=["send() method — two-way generators",
                      "itertools module overview: chain, cycle, islice, product"],
        practice=["Fibonacci Generator — infinite sequence, retrieve N terms",
                  "Log File Processor — stream-process a large file line-by-line"])

    week_header(8, "Modules, Packages & Project Structure", HexColor("#7C3AED"))

    add_class("22–23", "Modules, Packages & Virtual Environments", "90–120 min",
        concepts=["import, from import, as aliasing",
                  "Writing and organizing custom modules",
                  "Package structure: __init__.py and module hierarchy",
                  "Python's module search path (sys.path)",
                  "Virtual environments with venv — isolating dependencies",
                  "pip — installing, upgrading, pinning packages",
                  "requirements.txt — reproducible environments"],
        micro_topics=["__name__ == '__main__' guard",
                      "Relative vs. absolute imports",
                      "Namespace packages (Python 3.3+)"],
        practice=["Refactor Bank System into a proper package structure: bank/account.py, bank/utils.py"])

    add_class(24, "Module 2 Capstone — E-Commerce CLI", "120 min",
        concepts=["Integrating OOP, file I/O, exception handling, and modules",
                  "Designing a small system from scratch"],
        micro_topics=[],
        practice=["E-Commerce CLI System — Products, Cart, Orders, Checkout",
                  "Features: add/remove items, apply discount, persist cart to JSON",
                  "Use proper class design: Product, Cart, Order classes"],
        notes="Students present their systems. Code review + refactoring discussion.")

    story.append(PageBreak())

    # ══════════════════════════════════════════════════════════════════════════
    # MODULE 3
    # ══════════════════════════════════════════════════════════════════════════
    module_header(3, "Data Science Foundations", "9–12", 12, MINT)

    week_header(9, "NumPy — Numerical Computing", MINT)

    add_class("25–26", "NumPy Arrays & Vectorized Operations", "90–120 min",
        concepts=["Why NumPy? Performance gap between lists and arrays",
                  "ndarray — shape, dtype, dimensions (1D, 2D, 3D)",
                  "Array creation: np.array(), zeros(), ones(), arange(), linspace()",
                  "Indexing and slicing — rows, columns, regions",
                  "Broadcasting — operating arrays of different shapes",
                  "Vectorization — eliminating Python loops for speed"],
        micro_topics=["dtype precision: int32, float64, etc.",
                      "np.copy() vs. views — mutations and memory",
                      "Axis parameter — operations along rows (axis=0) or columns (axis=1)"],
        practice=["Numerical Analysis Tasks — mean, std, normalize a dataset, matrix multiply",
                  "Visualize broadcasting with small examples"])

    add_class(27, "NumPy — Statistical & Linear Algebra Tools", "90 min",
        concepts=["Aggregate functions: sum, mean, std, min, max, median",
                  "Boolean indexing — filter arrays by condition",
                  "np.where() — vectorized conditional logic",
                  "Random module: np.random for simulation",
                  "Intro to matrix operations (dot product, transpose)"],
        micro_topics=["np.nan handling: np.nanmean, np.isnan()",
                      "np.unique(), np.sort(), np.argsort()"],
        practice=["Grade Statistics Engine — compute class statistics from a score array",
                  "Random data simulation and analysis"])

    week_header(10, "Pandas — Data Manipulation", MINT)

    add_class("28–29", "Pandas DataFrames & Series", "90–120 min",
        concepts=["Series — 1D labeled array, indexed by default or custom index",
                  "DataFrame — 2D table, columns are Series",
                  "Loading data: read_csv(), read_excel(), read_json()",
                  "Exploring data: head(), tail(), info(), describe(), shape",
                  "Selecting columns, rows: [], loc[], iloc[]",
                  "Filtering rows with boolean conditions"],
        micro_topics=["Index alignment — how Pandas aligns by label",
                      "chained indexing warning — use .loc always",
                      "dtypes — understanding your column types"],
        practice=["Sales Dataset Analysis — load a CSV, explore structure, filter by region, sort by revenue"],
        table_headers=["Operation", "Method", "Use Case"],
        table_data=[["Load CSV", "read_csv('file.csv')", "Import tabular data"],
                    ["Preview", "df.head(10)", "First 10 rows"],
                    ["Summary", "df.describe()", "Statistical summary"],
                    ["Filter", "df[df['col'] > val]", "Conditional rows"],
                    ["Select col", "df['column']", "Single column Series"],
                    ["Slice rows", "df.iloc[0:5]", "Row range by position"],
                    ["Sort", "df.sort_values('col')", "Sort by column"]])

    add_class(30, "Pandas — GroupBy, Merge & Aggregation", "90 min",
        concepts=["groupby() — split-apply-combine pattern",
                  "Aggregation: sum, mean, count, min, max after groupby",
                  "Merging DataFrames: merge() (SQL-style joins)",
                  "Concatenating DataFrames: concat()",
                  "Pivot tables with pivot_table()"],
        micro_topics=["apply() — custom function over rows/columns",
                      "value_counts() — frequency distribution",
                      "crosstab() — relationship between two categorical columns"],
        practice=["Sales Dashboard — group by region, product, month; compute top performers"])

    week_header(11, "Data Cleaning & Visualization", MINT)

    add_class("31–32", "Data Cleaning with Pandas", "90–120 min",
        concepts=["Detecting missing values: isnull(), isna(), info()",
                  "Handling nulls: dropna(), fillna() with strategies (mean, ffill, constant)",
                  "Duplicate detection and removal: duplicated(), drop_duplicates()",
                  "Fixing dtypes: astype(), pd.to_datetime(), pd.to_numeric()",
                  "Renaming columns, resetting index",
                  "String cleaning: str.strip(), str.lower(), str.replace()"],
        micro_topics=["Outlier detection: IQR method with Pandas",
                      "Data validation: assert statements on cleaned data",
                      "Why clean data matters in ML — garbage in, garbage out"],
        practice=["Dirty Sales Dataset — identify and fix: nulls, duplicates, wrong dtypes, inconsistent categories"])

    add_class(33, "Matplotlib — Data Visualization", "90 min",
        concepts=["Matplotlib figure anatomy: Figure, Axes, Axis",
                  "Line plots — trends over time",
                  "Bar charts — categorical comparisons",
                  "Histograms — distribution of values",
                  "Scatter plots — relationships between variables",
                  "Customization: titles, labels, legends, colors, figsize"],
        micro_topics=["plt.subplots() — multiple plots in one figure",
                      "Seaborn as a higher-level wrapper (brief intro)",
                      "Saving figures: plt.savefig()"],
        practice=["Analysis Dashboard — 4-subplot figure: trend, category comparison, distribution, scatter",
                  "Style plots with custom colors matching a brand palette"])

    week_header(12, "End-to-End Data Science Pipeline", MINT)

    add_class("34–35", "Full Data Science Workflow", "90–120 min",
        concepts=["Step 1 — Data Collection: read from CSV / API / mock data",
                  "Step 2 — Exploration: shape, nulls, dtypes, describe()",
                  "Step 3 — Cleaning: handle nulls, fix types, remove duplicates",
                  "Step 4 — Analysis: groupby, aggregations, trends",
                  "Step 5 — Visualization: charts that tell a story",
                  "Step 6 — Insights: write findings as human-readable summary",
                  "Intro to ML: train/test split, what sklearn does (conceptual)"],
        micro_topics=["from sklearn.model_selection import train_test_split",
                      "What a model 'learns' — input → prediction",
                      "Overfitting intuition — why we test on unseen data"],
        practice=["Full Pipeline on a real dataset (e.g., student performance or retail sales)"])

    add_class(36, "Final Capstone — End-to-End Data Analysis Project", "120 min",
        concepts=["Students independently complete a full pipeline project",
                  "Dataset: chosen from Kaggle or provided by instructor",
                  "Deliverable: cleaned data + analysis + 4+ visualizations + written insights"],
        micro_topics=[],
        practice=["Capstone: E.g. 'Analyze Lahore Air Quality Data' or 'Sales Forecasting Report'",
                  "Students present findings — 5-minute demo each",
                  "Portfolio submission: Jupyter Notebook + PDF export"],
        notes="This is the final portfolio project. Encourage students to upload to GitHub.")

    story.append(PageBreak())

    # ══════════════════════════════════════════════════════════════════════════
    # LEARNING OUTCOMES + TOOLS
    # ══════════════════════════════════════════════════════════════════════════
    story.append(Paragraph("Learning Outcomes & Toolkit", S("OHead", fontName="Helvetica-Bold",
                            fontSize=16, textColor=VIOLET, spaceAfter=8, spaceBefore=4)))
    story.append(HRFlowable(width="100%", thickness=1, color=BORDER_LIGHT))
    story.append(Spacer(1, 10))

    outcomes_data = [
        ("Module 1 — Python Foundations",
         ["Write clean, readable Python following PEP 8",
          "Use all core data structures confidently: lists, dicts, sets, tuples",
          "Handle files, parse CSV & JSON, manage exceptions",
          "Build working CLI apps from scratch"]),
        ("Module 2 — Intermediate Python & OOP",
         ["Design object-oriented systems using classes, inheritance, polymorphism",
          "Write and apply decorators — logging, timing, access control",
          "Use generators for memory-efficient data processing",
          "Structure code as proper packages with virtual environments"]),
        ("Module 3 — Data Science Foundations",
         ["Perform vectorized operations with NumPy arrays",
          "Load, clean, filter, aggregate, and merge datasets with Pandas",
          "Build multi-chart visualizations with Matplotlib",
          "Execute a complete data science pipeline end-to-end",
          "Understand how ML models receive and process data"]),
    ]

    for mod_title, bullets in outcomes_data:
        out_data = [[
            Paragraph(mod_title, S("ot", fontName="Helvetica-Bold", fontSize=10,
                                   textColor=DEEP_VIOLET, leading=13)),
        ]]
        out_t = Table(out_data, colWidths=[W-36*mm])
        out_t.setStyle(TableStyle([
            ("BACKGROUND",    (0,0), (-1,-1), LIGHT_VIOLET),
            ("LEFTPADDING",   (0,0), (-1,-1), 12),
            ("TOPPADDING",    (0,0), (-1,-1), 6),
            ("BOTTOMPADDING", (0,0), (-1,-1), 6),
        ]))
        story.append(out_t)
        for b in bullets:
            story.append(Paragraph(f"✓  {b}", S("ob", fontName="Helvetica", fontSize=9,
                                                  textColor=TEXT_DARK, leftIndent=14,
                                                  spaceAfter=2, leading=12)))
        story.append(Spacer(1, 8))

    story.append(Spacer(1, 6))
    story.append(Paragraph("Tools & Technologies Used", S("TH", fontName="Helvetica-Bold",
                            fontSize=12, textColor=VIOLET, spaceAfter=6)))

    tools = [
        ["Language",     "Python 3.11+"],
        ["IDE",          "VS Code (recommended) / PyCharm"],
        ["Environment",  "venv / pip"],
        ["Data Science", "NumPy, Pandas, Matplotlib, Seaborn"],
        ["Version Ctrl", "Git + GitHub (encouraged from Week 4)"],
        ["Notebooks",    "Jupyter Notebook (Module 3)"],
        ["Data Sources", "Kaggle Datasets, UCI ML Repository"],
    ]

    tools_t = Table(tools, colWidths=[50*mm, W-36*mm-50*mm])
    tools_t.setStyle(TableStyle([
        ("FONTNAME",      (0,0), (0,-1), "Helvetica-Bold"),
        ("FONTNAME",      (1,0), (1,-1), "Helvetica"),
        ("FONTSIZE",      (0,0), (-1,-1), 9),
        ("TEXTCOLOR",     (0,0), (0,-1), DEEP_VIOLET),
        ("TEXTCOLOR",     (1,0), (1,-1), TEXT_DARK),
        ("ROWBACKGROUNDS",(0,0), (-1,-1), [ROW_ALT, white]),
        ("GRID",          (0,0), (-1,-1), 0.3, BORDER_LIGHT),
        ("TOPPADDING",    (0,0), (-1,-1), 5),
        ("BOTTOMPADDING", (0,0), (-1,-1), 5),
        ("LEFTPADDING",   (0,0), (-1,-1), 10),
        ("ROUNDEDCORNERS", [4]),
    ]))
    story.append(tools_t)

    story.append(Spacer(1, 16))
    story.append(HRFlowable(width="100%", thickness=0.5, color=BORDER_LIGHT))
    story.append(Spacer(1, 8))

    footer_text = (
        "© 2025 Quantum Logics Virtual Tech Academy  ·  Instructor: Abdul Aziz  ·  "
        "Rawalpindi, Pakistan  ·  All rights reserved."
    )
    story.append(Paragraph(footer_text, S("ft", fontName="Helvetica", fontSize=7,
                                           textColor=TEXT_LIGHT, alignment=TA_CENTER)))

    doc.build(story)
    print(f"PDF created: {OUTPUT}")


if __name__ == "__main__":
    make_doc()
```
